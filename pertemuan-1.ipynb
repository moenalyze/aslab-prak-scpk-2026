## 1. Variabel, Assignment, dan Tipe Data

Python tidak perlu deklarasi tipe data (`int`, `float`, `string`) secara eksplisit.
Untuk string, Python punya fitur f-string yang sangat memudahkan kita menyisipkan variabel ke dalam teks tanpa `<<` seperti di C++.

```python
x = "Hello, World" # string
x = 20 # integer
x = 20.5 # float
x = True # boolean
x = [1,2,3,4,5] # list
x = {"nama": "Budi", "umur": 25} # dictionary
```

Di C++, variabel adalah sebuah **kotak** yang menyimpan nilai.
Di Python, variabel hanyalah **label/tag** yang ditempelkan ke sebuah objek di memori. Mari kita buktikan dengan melihat alamat memorinya menggunakan fungsi `id()`.

```python
# Pembuktian Label vs Kotak
x = 'Informatika'
y = x  # y menempel pada objek yang sama dengan x

print(f"Isi x: {x} | Alamat x: {id(x)}")
print(f"Isi y: {y} | Alamat y: {id(y)}")
```

```python
print("\n-- Kita ubah x --")
x = 'Sisfo' # x menempel ke objek baru, y tetap di tempat lama

print(f"Isi x: {x} | Alamat x: {id(x)}")
print(f"Isi y: {y} | Alamat y: {id(y)}")
```

```python
# F-String (Formatted String)
name = "Khatama Putra"
print(f"Hello, nama saya {name}")
```

```python
# Multiline String
multi_line = """Halo!
Kapan terakhir kali kita bertemu?
Kita bertemu hari Jum’at yang lalu."""
print(multi_line)
```

## 2. List

Berbeda dengan Array di C++ yang statis, `List` di Python bersifat dinamis dan bisa diisi tipe data campuran. Kita juga bisa "memotong" (slicing) List dengan mudah menggunakan sintaks `[start:stop:step]`.

```python
x = [1, 2.2, 'Dicoding']
print(type(x))
print(x[2])

x[2] = 'Indonesia' # mutable
print(x[2])
```

```python
x = ["laptop", "monitor", "mouse", "mousepad", "keyboard", "webcam", "microphone"]
```

```python
# Indexing
print("Barang pertama:", x[0])
print("Barang terakhir:", x[-1])
```

```python
# Slicing [start:stop:step] (Stop bersifat eksklusif / tidak ikut)
print("3 Barang pertama:", x[:3])
print("Lompat 2 langkah:", x[0:5:2])
```

```python
# Cek keberadaan elemen di list
print('mouse' in x)
print('komputer' in x)
print('microphone' not in x)
print('meja' not in x)
```

```python
# Built-in function untuk perhitungan
angka = [13, 7, 24, 5, 96]
print("Nilai Min:", min(angka))
print("Nilai Max:", max(angka))
print("Jumlah data:", len(angka))
```

```python
kendaraan = ['motor', 'mobil', 'helikopter', 'pesawat']
kendaraan.sort() # mengurutkan ascending
print(kendaraan)

kendaraan.sort(reverse=True) # mengurutkan descending
print(kendaraan)
```

```python
# Kapital akan tercetak lebih dulu karena pakai ASCII
kendaraan = ['motor', 'mobil', 'helikopter', 'Pesawat']
kendaraan.sort()

print(kendaraan)
```

```python
# Tidak bisa mengurutkan jika tipe data campuran
urutan = ['UPN', 1, 2, 'Jogja', 3]
urutan.sort() # TypeError: '<' not supported between instances of 'int' and 'str'
```

## 4. Operator

Operator aritmatika

```python
x = 3
y = 5

print('Penjumlahan\t: ', x + y)
print('Pengurangan\t: ', x - y)
print('Perkalian\t: ', x * y)
print('Pembagian\t: ', x / y)
print('Modulus\t\t: ', x % y)
print('Eksponen\t: ', x ** y)
print('Pembagian bulat\t: ', x // y)
```

Operator logika

1. and: Akan menghasilkan nilai true jika kedua statement memiliki nilai true dan false untuk lainnya.
2. or: Akan menghasilkan nilai true jika salah satu statement memiliki nilai true.
3. not: Membalikkan nilai statement, jika false akan menjadi true.

Contoh lebih jelas ada pada bagian statement pengendalian.

```python
i = 5

# Penulisan C++
# i++; # error

# Penulisan Python
i += 1
```

## 5. Input/Output

```python
matematika = int(input('Matematika: '))
inggris = int(input('B. Inggris: '))
ipa = int(input('IPA: '))

ratarata = (matematika + inggris + ipa)/3

if ratarata > 90:
    print('Nilai: ', ratarata)
    print('Sangat Baik')
elif ratarata > 80:
    print('Nilai: ', ratarata)
    print('Baik')
elif ratarata > 70:
    print('Nilai: ', ratarata)
    print('Cukup')
else:
    print('Nilai: ', ratarata)
    print('Perlu Perbaikan')
```

## 7. Looping

```python
# Print 0 sampai 2
count = 0
while count < 3:
  print(count)
  count += 1
```

```python
# Print 0 sampai 9
for i in range(10):
    print(i)
```

```python
# Print 1 sampai 5
for i in range(1,6):
  print(i)
```

```python
# Print 1 sampai 9, lompat 2 langkah
for i in range(1,10,2): # (start,end,step)
    print(i)
```

```python
penyanyi = ['Nadin Amizah', 'Clairo', 'Olivia Dean', 'Laufey']

for nama in penyanyi:
    print(f'Halo {nama}')
```

```python
# Print dengan indeks
# Cara 1
# for i in range(len(penyanyi)):
#     print(f"Penyanyi ke-{i}: {penyanyi[i]}")
  
# Cara 2 (prefer)
for i, nama in enumerate(penyanyi, start=1):
    print(f"Penyanyi ke-{i}: {nama}")
```

Bikin list baru dari list yang sudah ada

```python
# Cara 1
angka = [1, 2, 3, 4]
hasil = []
for x in angka:
    hasil.append(x ** 2)
print(hasil)
```

```python
# Cara 2 (prefer)
angka = [1, 2, 3, 4]
hasil = [x ** 2 for x in angka]
print(hasil)
```

```python
# bikin list yang belum ada
var_arr = [i for i in range(10)]
print(var_arr)
```

Tambahan: for-else

```python
numbers = [1, 2, 3, 4, 5]

for num in numbers:
    if num == 6:
        print("Angka ditemukan! Program berhenti!")
        break
else:
    print("Angka tidak ditemukan.")
```

```python
# Kalau else-nya sejajar if jadinya berulang terus
for num in numbers:
    if num == 6:
        print("Angka ditemukan! Program berhenti!")
        break
    else:
        print("Angka tidak ditemukan.")
```
