# WP - Pemilihan HP

Kriteria:
1. Harga (Juta Rp) -> Cost (Makin murah makin bagus)
2. RAM (GB) -> Benefit (Makin besar makin bagus)
3. Kamera (MP) -> Benefit (Makin besar makin bagus)
4. Baterai (mAh) -> Benefit (Makin besar makin bagus)

Alternatif:
1. Samsung (A1): Rp 4 Juta, RAM 8GB, Kamera 50MP, Baterai 5000mAh
2. Xiaomi (A2): Rp 3 Juta, RAM 12GB, Kamera 108MP, Baterai 4500mAh
3. iPhone (A3): Rp 10 Juta, RAM 6GB, Kamera 12MP, Baterai 3500mAh

Bobot (Skala 1 - 5):
1. Harga: 5
2. RAM: 4
3. Kamera: 3
4. Baterai: 4

## WP (Sesuai Modul)

```python
import numpy as np
import pandas as pd
```

```python
alternatif = ['Samsung', 'Xiaomi', 'iPhone']
kriteria = ['Harga', 'RAM', 'Kamera', 'Baterai']

# Matriks Keputusan (X)
# Baris = Alternatif, Kolom = Kriteria (Harga, RAM, Kamera, Baterai)
data = np.array([
    [4,  8,  50, 5000],  # A1
    [3, 12, 108, 4500],  # A2
    [10, 6,  12, 3500]   # A3
])

# Bobot Preferensi Awal (Menggunakan Skala 1 - 5)
# 1: Sangat Tidak Penting, 2: Tidak Penting, 3: Cukup Penting, 4: Penting, 5: Sangat Penting
# Urutan: [Harga (5), RAM (4), Kamera (3), Baterai (4)]
bobot_awal = np.array([5, 4, 3, 4])

# (-1 = Cost, 1 = Benefit)
k = np.array([-1, 1, 1, 1])

df = pd.DataFrame(data, index=alternatif, columns=kriteria)
print("=== Matriks Keputusan Awal ===")
print(df)
print("\nBobot Awal (Skala 1-5):", bobot_awal)
```

### Normalisasi Nilai Bobot

```python
# m = len(data)
# n = len(data[0])
# norm_bobot = [b / sum(bobot) for b in bobot]

# Alternatif kode
m, n = data.shape
norm_bobot = bobot_awal / np.sum(bobot_awal)

print("Bobot Ternormalisasi:", norm_bobot)
```

### Menghitung Nilai Preferensi (Vektor S)

```python
s = []

for i in range(m):
  s_value = 1
  for j in range(n):
    s_value *= data[i][j] ** (k[j] * norm_bobot[j])
  s.append(s_value)
  
# Alternatif kode
s = np.prod(data ** (k * norm_bobot), axis=1)
  
s_df = pd.DataFrame(s, index=alternatif, columns=["Vektor S"])
print(s_df)
```

### Normalisasi Vektor S (Vektor V)

```python
v = s / np.sum(s)

v_df = pd.DataFrame(v, index=alternatif, columns=["Vektor V"])
sort_ranking = v_df.sort_values(by='Vektor V', ascending=False)
print("=== Hasil Ranking ===")
print(sort_ranking)
print()

best_alternative = sort_ranking.index[0]
print(f"Alternatif terbaik adalah {best_alternative}")
```

## WP (PPT Bu Herlina)

```python
import numpy as np
import pandas as pd
```

```python
alternatif = ['Samsung', 'Xiaomi', 'iPhone']
kriteria = ['Harga', 'RAM', 'Kamera', 'Baterai']

# Matriks Keputusan (X)
# Baris = Alternatif, Kolom = Kriteria (Harga, RAM, Kamera, Baterai)
data = np.array([
    [4,  8,  50, 5000],  # A1
    [3, 12, 108, 4500],  # A2
    [10, 6,  12, 3500]   # A3
])

# Bobotnya sudah ternormalisasi (bukan skala)
bobot = np.array([0.30, 0.20, 0.20, 0.30])

# (0 = Cost, 1 = Benefit)
atribut = np.array([0, 1, 1, 1])

df = pd.DataFrame(data, index=alternatif, columns=kriteria)
print("=== Matriks Keputusan Awal ===")
print(df)
print("\nBobot Awal (Skala 1-5):", bobot)
```

### Normalisasi Matriks Keputusan (Seperti SAW)

```python
m, n = data.shape
normalized_matrix = np.zeros((m,n))

for j in range(n):
    if atribut[j] == 1: # Kriteria Benefit
        normalized_matrix[:, j] = data[:, j] / np.max(data[:, j])
    else: # Kriteria Cost
        normalized_matrix[:, j] = np.min(data[:, j]) / data[:, j]
    
normalized_df = pd.DataFrame(normalized_matrix, index=alternatif, columns=kriteria)
print('=== Matriks Ternormalisasi (ala SAW) ===')
print(normalized_df)
```

### Menghitung Skor Kinerja & Ranking

```python
skor_kinerja = []

for i in range(m):
    skor = 1
    for j in range(n):
        # Semua pangkat POSITIF karena atribut cost/benefit sudah diselesaikan saat normalisasi matriks
        skor *= normalized_matrix[i][j] ** bobot[j]
    skor_kinerja.append(skor)

skor_df = pd.DataFrame(skor_kinerja, index=alternatif, columns=["Skor Kinerja WP"])
sort_ranking = skor_df.sort_values(by='Skor Kinerja WP', ascending=False)

print("\n=== Hasil Ranking (WP Modifikasi Teori) ===")
print(sort_ranking)
print(f"\nAlternatif Terbaik adalah: {sort_ranking.index[0]}")
```
