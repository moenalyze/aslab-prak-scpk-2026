# AHP - Pemilihan Motor

Penentuan **Alternatif** dan **Kriteria**
- Alternatif:
  - A1: Yamaha
  - A2: Honda
  - A3: Suzuki
  - A4: Kawasaki
- Kriteria:
  - C1: Gaya
  - C2: Keandalan
  - C3: Keekonomisan Bahan Bakar

**Tingkat Kepentingan**:
| Intensitas | Keterangan |
|------------|------------|
| 1 | X dan Y sama penting/baik |
| 2 | X sedikit lebih penting/baik dari Y |
| 3 | X lebih penting/baik dari Y |
| 4 | X jauh lebih penting/baik dari Y |

**Preferensi Kriteria**
1. *Keandalan* dua kali lebih penting dari gaya.
2. *Gaya* tiga kali lebih penting dari keekonomisan bahan bakar.
3. *Keandalan* empat kali lebih penting dari keekonomisan bahan bakar.

**Preferensi Alternatif**
1. Preferensi Gaya:
   - Kawasaki > Honda > Yamaha > Suzuki
2. Preferensi Keandalan:
   - Honda > Yamaha > Kawasaki > Suzuki 
3. Nilai Keekonomisan Bahan Bakar:
   - Yamaha  : 60
   - Honda   : 80
   - Suzuki  : 60
   - Kawasaki: 80

```python
import numpy as np
```

```python
def normalize_comparation(M):
    """Fungsi untuk normalisasi matriks perbandingan berpasangan"""
    # menjumlahkan elemen-elemen dalam kolom yang sama
    sumCol = M.sum(axis=0) # axis 0 = kolom, 1 = baris
    norm = M / sumCol
    return norm

def weight(M):
    """Fungsi untuk menghitung bobot"""
    # menghitung rata-rata baris
    return np.mean(M, axis=1)

def validity_check(M, W):
    """Fungsi untuk mengecek validitas bobot"""
    n = len(M)
    RI = {
        2: 0.00, 3: 0.58, 4: 0.90, 5: 1.12, 6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45,
        10: 1.51, 11: 1.53, 12: 1.54, 13: 1.56, 14: 1.57
    }
    
    # menghitung CV
    CV = M @ W / W # operator @ mengalikan matriks sesuai dengan prinsipnya
    print(f"CV: {CV}\n")
    
    # menghitung nilai eigen
    eigen = np.mean(CV)
    print(f"Eigen: {eigen}\n")
    
    # mencari nilai RI
    print(f"RI: {RI[n]}\n")
    
    # menghitung CI
    CI = (eigen - n) / (n - 1)
    print(f"CI: {CI}\n")
    
    # menghitung CR
    CR = CI / RI[n]
    print(f"CR: {CR}\n")
    
    if CR <= 0.1:
        print("Bobot valid!\n")
    else:
        print("Bobot tidak valid!\n")

def final_weight(W_alt, W_crit):
    """Menghitung bobot akhir tiap alternatif"""
    # mengalikan bobot alternatif dan kriteria sesuai dengan prinsip matriks
    W_fin = W_alt.T @ W_crit
    return W_fin
```

## 1. Evaluasi Bobot Kriteria

Matriks Perbandingan Berpasangan Kriteria

```python
# alternatif
A = ["Yamaha", "Honda", "Suzuki", "Kawasaki"]

print("=== Kriteria ===")
# matriks perbandingan berpasangan kriteria
# urutan: gaya, keandalan, keekonomisan
MPBk = np.array([
    [1, 1/2, 3],
    [2, 1, 4],
    [1/3, 1/4, 1]
])
print(f"Matriks perbandingan:\n{MPBk}")
```

Normalisasi & Perhitungan Bobot

```python
# normalisasi
MPBk_norm = normalize_comparation(MPBk)
print(f"Matriks perbandingan setelah normalisasi:\n{MPBk_norm}")

# perhitungan bobot
Wk = weight(MPBk_norm)
print(f"\nBobot kriteria:\n{Wk}")
```

Uji Validitas Bobot

```python
# uji validitas bobot
validity_check(MPBk, Wk)
```

## 2. Evaluasi Alternatif Tiap Kriteria

### Kriteria Gaya

```python
print("=== Kriteria: Gaya ===")
# urutan: yamaha, honda, suzuki, kawasaki
MPBg = np.array([
    [1, 1/2, 2, 1/3],
    [2, 1, 3, 1/2],
    [1/2, 1/3, 1, 1/4],
    [3, 2, 4, 1]
])
MPBg_norm = normalize_comparation(MPBg)
Wg = weight(MPBg_norm)
print(f"Bobot Gaya (Wg): {Wg}\n")
validity_check(MPBg, Wg)
```

### Kriteria Keandalan

```python
print("=== Kriteria: Keandalan ===")
MPBa = np.array([
    [1, 1/2, 3, 2],
    [2, 1, 4, 3],
    [1/3, 1/4, 1, 1/2],
    [1/2, 1/3, 2, 1]
])
MPBa_norm = normalize_comparation(MPBa)
Wa = weight(MPBa_norm)
print(f"Bobot Keandalan (Wa): {Wa}\n")
validity_check(MPBa, Wa)
```

### Kriteria Keekonomisan

```python
print("=== Kriteria: Keekonomisan ===")
MPBe = np.array([
    [1, 60/80, 1, 60/80],
    [80/60, 1, 80/60, 1],
    [1, 60/80, 1, 60/80],
    [80/60, 1, 80/60, 1]
])
MPBe_norm = normalize_comparation(MPBe)
We = weight(MPBe_norm)
print(f"Bobot Keekonomisan (We): {We}\n")
validity_check(MPBe, We)
```

## 3. Perankingan Akhir

```python
# menyatukan semua bobot alternatif dalam satu matriks (Rekap Bobot)
W_total = np.array([Wg, Wa, We])
print(f"Matriks Rekap Bobot Alternatif (W_total):\n{W_total}\n")

# menghitung bobot akhir
W_final = final_weight(W_total, Wk)
print(f"Bobot akhir tiap alternatif:\n{W_final}\n")

# memunculkan alternatif terbaik
max_index = np.argmax(W_final)
print(f"Alternatif terbaik adalah {A[max_index]} dengan nilai {W_final[max_index]:.4f}")
```
