## Styling Grafik

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1, 2, 3, 4])
y = x**2

plt.title(
    'Grafik Sederhana', # judul grafik
    fontsize = 20, # ukuran font
    fontstyle = 'italic', # gaya font
    family = 'Times New Roman', # jenis font family
    # font = 'Times New Roman', # hindari karena akan mereset properti lain
    # color = 'red', # warna font pake bahasa inggris
    # color = 'r', # warna font pake singkatan
    # color = '#16A985', # warna font pake HEX
    color = (0.5, 0.2, 0.8),
    pad = 20 # jarak judul dengan kotak grafik
)
plt.xlabel(
    "Sumbu X",
    color = 'b', # warna label sumbu x
    # fontweight = 'bold', # ketebalan
    fontweight = 700, # ketebalan
    labelpad = 25 # jarak label dari sumbu x
)
plt.ylabel(
	"Sumbu Y",
	color = 'b',
	fontweight = "bold",
	labelpad = 25
)
plt.plot(
	x, y,
	marker='o',
	color = "#42eff5",
	#   linestyle = "--", # gaya garis putus-putus
	linestyle = "-.", # gaya garis putus-titik
	linewidth = 2, # lebar garis
	markersize = 5 # ukuran marker
)
# plt.plot(x, y, 'b-.o') # [color][linestyle][marker] (urutannya dibolak-balik bebas)
plt.show()
```

### Menambahkan Grid

```python
x = np.array([1, 2, 3, 4])
y = x**2

plt.title(
    'Grafik Sederhana', # judul grafik
    fontsize = 20, # ukuran font
    fontstyle = 'italic', # gaya font
    family = 'Times New Roman', # jenis font family
    # font = 'Times New Roman', # hindari karena akan mereset properti lain
    # color = 'red', # warna font pake bahasa inggris
    # color = 'r', # warna font pake singkatan
    # color = '#16A985', # warna font pake HEX
    color = (0.5, 0.2, 0.8),
    pad = 20 # jarak judul dengan kotak grafik
)
plt.xlabel(
    "Sumbu X",
    color = 'b', # warna label sumbu x
    # fontweight = 'bold', # ketebalan
    fontweight = 700, # ketebalan
    labelpad = 25 # jarak label dari sumbu x
)
plt.ylabel(
	"Sumbu Y",
	color = 'b',
	fontweight = "bold",
	labelpad = 25
)
plt.grid( # menambahkan grid
    True,
    linestyle = '-.',
    alpha = 0.3,
    linewidth = 2
)
plt.plot(
	x, y,
	marker='o',
	color = "#42eff5",
	#   linestyle = "--", # gaya garis putus-putus
	linestyle = "-.", # gaya garis putus-titik
	linewidth = 2, # lebar garis
	markersize = 5 # ukuran marker
)
# plt.plot(x, y, 'b-.o') # [color][linestyle][marker] (urutannya dibolak-balik bebas)
plt.show()
```

### Menambahkan Legend

```python
x = np.array([1, 2, 3, 4])  
y1 = x**2
y2 = x**3
y3 = [1, 5, 10, 20]

plt.title(
    'Grafik Sederhana', # judul grafik
    fontsize = 20, # ukuran font
    fontstyle = 'italic', # gaya font
    family = 'Times New Roman', # jenis font family
    # font = 'Times New Roman', # hindari karena akan mereset properti lain
    # color = 'red', # warna font pake bahasa inggris
    # color = 'r', # warna font pake singkatan
    # color = '#16A985', # warna font pake HEX
    color = (0.5, 0.2, 0.8),
    pad = 20 # jarak judul dengan kotak grafik
)
plt.xlabel(
    "Sumbu X",
    color = 'b', # warna label sumbu x
    # fontweight = 'bold', # ketebalan
    fontweight = 700, # ketebalan
    labelpad = 25 # jarak label dari sumbu x
)
plt.ylabel(
	"Sumbu Y",
	color = 'b',
	fontweight = "bold",
	labelpad = 25
)
plt.grid( # menambahkan grid
    True,
    linestyle = '-.',
    alpha = 0.3,
    linewidth = 2
)
plt.plot(
	x, y,
	marker='o',
	color = "#42eff5",
	#   linestyle = "--", # gaya garis putus-putus
	linestyle = "-.", # gaya garis putus-titik
	linewidth = 2, # lebar garis
	markersize = 5, # ukuran marker
	label = '$y_1 = x^2$'
)
plt.plot(
	x, y2,
	marker='o',
	color = "#7542f5",
	#   linestyle = "--", # gaya garis putus-putus
	linestyle = "-.", # gaya garis putus-titik
	linewidth = 2, # lebar garis
	markersize = 5, # ukuran marker
 	label = '$y_2 = x^3$'
)
plt.plot(
	x, y3,
	marker='o',
	color = "#f54296",
	#   linestyle = "--", # gaya garis putus-putus
	linestyle = "-.", # gaya garis putus-titik
	linewidth = 2, # lebar garis
	markersize = 5, # ukuran marker
 	label = '$y_3 = [1, 5, 10, 20]$'
)
# plt.plot(x, y, 'b-.o') # [color][linestyle][marker] (urutannya dibolak-balik bebas)
plt.legend(
	loc = 'upper left', # lokasi kiri atas
    fontsize = 12,      # ukuran huruf di dalam kotak legend
    shadow = True,      # ngasih efek bayangan 3D di kotaknya
    # ncol = 3            # menyusun labelnya jadi 3 kolom menyamping (defaultnya ke bawah)
    bbox_to_anchor=(1.0, 1.0)    # koordinat (X, Y) di luar grafik. 1.0 artinya di ujung kanan.
)
plt.show()
```

## Grafik Fungsi Matematika

`np.arange(start, stop, step)`

- start (Titik Awal): Angka pertama yang akan dimasukkan ke dalam array. (Kalau tidak ditulis, otomatis dimulai dari 0).

- stop (Titik Henti): Angka batas akhir. PENTING: Angka ini sifatnya eksklusif (tidak ikut dimasukkan ke dalam array). Program akan berhenti tepat sebelum menyentuh angka ini.

- step (Lompatan): Jarak antar angka. (Kalau tidak ditulis, otomatis jaraknya 1).

```python
# Bikin array dari 0 sampai 4 (5-nya nggak ikut)
a = np.arange(5)
print(a) 
```

```python
# Bikin array dari 2 sampai 9
b = np.arange(2, 10)
print(b)
```

```python
# Bikin array dari -3 sampai 3, dengan lompatan 0.5
c = np.arange(-3, 3.1, 0.5)
print(c)
```

$$Domain = \{x \in \mathbb{R} \mid -3 \le x \le 3\}$$

```python
x = np.arange(-3, 3.1, 0.1)  # Inkremen 0.1 agar grafik terlihat mulus
y = x ** 3
plt.plot(x, y)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Kurva $y = x^3$')
plt.grid(True)
plt.show()
```

```python
x = np.arange(-3, 3.1, 0.1)  # Inkremen 0.1 agar grafik terlihat mulus
y = x ** 3
plt.plot(x, y, label='x')
plt.plot(x + 5, y, label='x + 5')
plt.plot(x + 10, y, label='x + 10')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Kurva $y = x^3$')
plt.grid(True)
plt.legend()
plt.show()
```

## Load & Visualisasi Data CSV

```python
import pandas as pd
```

```python
df = pd.read_csv('data_hp.csv')
display(df.head())
```

### Scatter Plot

```python
# Visualisasi: Hubungan Resolusi Kamera dan Harga
plt.figure(figsize=(8, 5))
plt.scatter(df['Kamera_MP'], df['Harga_Juta'], color='red', marker='o')

plt.title("Hubungan Resolusi Kamera (MP) vs Harga HP")
plt.xlabel("Resolusi Kamera (MP)")
plt.ylabel("Harga (Juta Rp)")
plt.grid(True, linestyle='--')
plt.show()
```

### Bar Plot

```python
# Visualisasi: Total Penjualan per HP
plt.figure(figsize=(10, 5))
plt.bar(df['ID_HP'], df['Terjual_Bulan_Ini'], color='skyblue')

plt.title("Total Penjualan Smartphone Bulan Ini")
plt.xlabel("ID Smartphone")
plt.ylabel("Jumlah Terjual (Unit)")
plt.show()
```

### Pie Chart

```python
# Menghitung total penjualan berdasarkan "Merek"
penjualan_merek = df.groupby('Merek')['Terjual_Bulan_Ini'].sum()

# Visualisasi: Market Share (Pangsa Pasar)
plt.figure(figsize=(7, 7))
plt.pie(penjualan_merek, labels=penjualan_merek.index, autopct='%1.1f%%', startangle=90)

plt.title("Persentase Penjualan Berdasarkan Merek")
plt.show()
```
