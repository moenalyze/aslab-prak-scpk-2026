# SAW - Pemilihan HP

Kriteria:
1. Harga (Juta Rp) -> Cost (Makin murah makin bagus)
2. RAM (GB) -> Benefit (Makin besar makin bagus)
3. Kamera (MP) -> Benefit (Makin besar makin bagus)
4. Baterai (mAh) -> Benefit (Makin besar makin bagus)

Alternatif:
1. Samsung (A1): Rp 4 Juta, RAM 8GB, Kamera 50MP, Baterai 5000mAh
2. Xiaomi (A2): Rp 3 Juta, RAM 12GB, Kamera 108MP, Baterai 4500mAh
3. iPhone (A3): Rp 10 Juta, RAM 6GB, Kamera 12MP, Baterai 3500mAh

## Menggunakan Data Manual

```python
import numpy as np
import pandas as pd
```

Alternatif

```python
alternatif = ['Samsung', 'Xiaomi', 'iPhone']
kriteria = ['Harga', 'RAM', 'Kamera', 'Baterai']

# Matriks Keputusan (X)
# Baris = Alternatif, Kolom = Kriteria (Harga, RAM, Kamera, Baterai)
data = np.array([
    [4,  8,  50, 5000],  # A1
    [3, 12, 108, 4500],  # A2
    [10, 6,  12, 3500]   # A3
])

# Bobot Preferensi (W) - Total harus 1.0 (atau 100%)
# Urutan: [Harga, RAM, Kamera, Baterai]
bobot = np.array([0.30, 0.20, 0.20, 0.30])

# Menentukan Nilai Atribut masing-masing Kriteria
# 0 = Cost, 1 = Benefit
atribut = np.array([0, 1, 1, 1])

df = pd.DataFrame(data, index=alternatif, columns=kriteria)
print('Matriks Keputusan :')
print(df)

print('\nBobot yang diberikan')
print('Bobot = ', bobot)
print('\nAtribut tiap kriteria')
print('Atribut = ', atribut)
```

TAHAP 1: Normalisasi

```python
m, n = data.shape
normalized_matrix = np.zeros((m,n))

# Normalisasi
for j in range(n):
  if atribut[j] == 1: # Kriteria Benefit
    normalized_matrix[:, j] = data[:,j] / np.max(data[:,j])
  else: # Kriteria Cost
    normalized_matrix[:, j] = np.min(data[:,j]) / data[:,j]
    
normalized_df = pd.DataFrame(normalized_matrix, index=alternatif, columns=kriteria)
print('Matriks Ternormalisasi :')
print(normalized_df)
```

TAHAP 2: Menghitung Nilai Preferensi

```python
value_preferensi = np.dot(normalized_matrix, bobot) # Ngedot / Perkalian matriks

# Hasil
ranking = pd.DataFrame(value_preferensi, index=alternatif, columns=['Nilai Preferensi'])
sort_ranking = ranking.sort_values(by='Nilai Preferensi', ascending=False)
print('\nHasil Ranking :')
print(sort_ranking)

best_alternatif = sort_ranking.index[0]
print(f'\nAlternatif Terbaik adalah : {best_alternatif}')
```

## Menggunakan Dataset CSV

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('smartphone.csv')

print("=== Data Awal ===")
print(df)
```

```python
alternatif = df.iloc[:, 0].tolist()  # Mengambil kolom alternatif
kriteria = df.columns[1:].tolist() # Mengambil kolom kriteria

data = df.iloc[:, 1:].values     # Mengambil semua kolom angka

# Bobot Preferensi (W) - Total harus 1.0 (atau 100%)
# Urutan: [Harga, RAM, Kamera, Baterai]
bobot = np.array([0.30, 0.20, 0.20, 0.30])

# Menentukan Nilai Atribut masing-masing Kriteria
# 0 = Cost, 1 = Benefit
atribut = np.array([0, 1, 1, 1])

print('alternatif:\n', alternatif)
print('kriteria:\n', kriteria)
print('data:\n', data)
```

TAHAP 1: Normalisasi

```python
m, n = data.shape
normalized_matrix = np.zeros((m,n))

# Normalisasi
for j in range(n):
  if atribut[j] == 1: # Data Benefit
    normalized_matrix[:, j] = data[:,j] / np.max(data[:,j])
  else: # Data Cost
    normalized_matrix[:, j] = np.min(data[:,j]) / data[:,j]
    
normalized_df = pd.DataFrame(normalized_matrix, index=alternatif, columns=kriteria)
print('Matriks Ternormalisasi :')
print(normalized_df)
```

TAHAP 2: Menghitung Nilai Preferensi

```python
value_preferensi = np.dot(normalized_matrix, bobot) # Ngedot / Perkalian matriks

# Hasil
ranking = pd.DataFrame(value_preferensi, index=alternatif, columns=['Nilai Preferensi'])
sort_ranking = ranking.sort_values(by='Nilai Preferensi', ascending=False)
print('\nHasil Ranking :')
print(sort_ranking)

best_alternatif = sort_ranking.index[0]
print(f'\nAlternatif Terbaik adalah : {best_alternatif}')
```
